```python
plt.scatter(df["Sleep Duration"] , df["Stress Level"])
plt.title("LIFESTYLE")
plt.xlabel("SLEEP DURATION")
plt.ylabel("STRESS LEVEL")
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='pie',autopct ='%f%%' )
plt.show()
```

```python
df['Occupation'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.plot(df.groupby("Heart Rate")["Physical Activity Level"].mean(),marker = 'o')
plt.title("LIFESTYLE")
plt.xlabel("PHYSICAL ACTIVITY")
plt.ylabel("HEART RATE")
plt.show()
```
